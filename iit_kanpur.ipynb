```python
import numpy as np

def calculate_stats(data):
    sample_size = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    return sample_size, mean, std_dev
```

```python
from google.colab import files
uploaded=files.upload()
df=pd.read_csv('hsb2 (2).csv')

for col in ['read', 'write', 'math', 'science','id','race','ses','schtyp','prog','socst','Gender']:
    data = df[col].head(201)
    sample_size, mean, std_dev = calculate_stats(data)
    print(f"Statistics for {col}:")
    print(f"  Sample Size: {sample_size}")
    print(f"  Mean: {mean}")
    print(f"  Standard Deviation: {std_dev}")
    print("-" * 20)
```

```python
df = pd.read_csv('hsb2 (2).csv')

df['total'] = df[['read', 'write', 'math', 'science','socst']].head(9).sum(axis=1)

print(df.head(201))
```

```python

df['total'] = df['read'] + df['write'] + df['math'] + df['science'] + df['socst']

total_mean = df['total'].mean()
total_std = df['total'].std()

print(f"Mean of total marks: {total_mean}")
print(f"Standard deviation of total marks: {total_std}")
```
